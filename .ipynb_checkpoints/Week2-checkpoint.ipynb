```python
# How to plot it
sns.set_theme(style="white")

# Compute the correlation matrix
corr = data.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool))

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(7, 5))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```

## Outliers

### What's `Outliers`?
**ELI5:** Numbers parts of a dataset very different from all other numbers.

**Also,** In stats, scientists notice outliers because they might be mistakes or they might be important to pay attention to. The boxplot is made with the `IQR`

**MORE...** For example, if all your friends are 5 feet tall, but one person is 6 feet tall, they are the outlier because they are much taller than everyone else.

```python
# How to plot it

# To show the outliers of a column, we have to select them
sns.boxplot(x=data['IC1'])
```

### What's  `IQR`?
**ELI5:** Inter-Quartile Range.

**Also,** In stats, scientists notice outliers because they might be mistakes or they might be important to pay attention to. The boxplot is made with the `IQR`

**MORE...** Quartiles split the dataset in 4 equal parts, each containing 25% of the observations. Between the first and third quartile are the middle 50% of the observations. Hence, the minimum and maximum of the middle 50% are identical with Q1 and Q3. Therefore, these definitions are equivalent and should lead to the same IQR for the same data.



```python
# How to remove the Outliers
sns.boxplot(x=data['IC1'], whis=5.7) # By default whis=1.5. We can change it to expand the range and remove outliers.
```

```python
# Interval of Q1
np.percentile(data['IC1'], 25)
```

```python
# Interval of Q3
np.percentile(data['IC1'], 75)
```

```python
# IQR is the difference between Q3 - Q1
iqr = np.percentile(data['IC1'], 75) - np.percentile(data['IC1'], 25)
iqr
```

```python
# Defining the limits
upper_limit = np.percentile(data['IC1'], 75) + 5.7*iqr # We use 5.7 as we did to remove outliers. The standard is 1.5
lower_limit = np.percentile(data['IC1'], 25) - 5.7*iqr # Same here
upper_limit, lower_limit
```

```python
# We check again the data for IC1. Is the min a negative number?
data.describe()
```

```python
# Data before erasing
datat = data.shape
datat
```

```python
# BEFORE REMOVING THE DATA: How many rows are the Outliers?
len(data[(data['IC1']<lower_limit) & (data['IC1']>upper_limit)])
```

```python
# Now, so we have to fix the lower_limit to erase negative numbers.
data = data[(data['IC1']>lower_limit) & (data['IC1']<upper_limit)]
data
```

### What's a `probability density function`?

**ELI5:** How likely something is to happen. 
When we look at the graph, the highest point is the most likely thing to happen, and the lowest point is the least likely thing to happen. The area under the graph shows us how likely it is for something to happen.

**Also,** A probability density function is used where the possibilities are `continuous`, not discrete. (Like the `normal distribution`).

```python
# How to plot it

# create a subplot with 1 figure and 1 axis
fig, ax = plt.subplots(1,1)

# We store within a interval (-5 to 5) evenly spaced numbers (400)
x = np.linspace(-5,5,400)

from scipy.stats import norm # method from scipy library
# We use the Probability Density (↑ check one-cell above) Function on X
y= norm.pdf(x) 

# How the axis in the plot will be distributed
ax.plot(x, y, 'b+') #plot x and y using blue plusses
plt.show()
```

### What's `Mean Squared Error`?

**ELI5:** How well a model fits the datapoins.

**Also,** when we draw a line or use a model to predict a value, the mean squared error tells us how close the predicted values are to the actual data points.

**MORE...** It is calculated by taking the difference between the actual value and the predicted value, squaring the difference (to make it positive, since the differences could be negative or positive), and then taking the average of all these squared differences. A lower mean squared error means the line or model fits the data better.



### What's `Coefficient of determination`?

**ELI5:** How well a prediction matches up with the actual data.

**Also,** The higher the score the better.

**MORE...** A scientist would compare their prediction to the actual data and then get a number from 0-1. A number of 0 means the prediction was no better than a guess and a number of 1 means the prediction perfectly matched the data.



```python
from sklearn.metrics import mean_squared_error, r2_score # These two matrix, "Mean Squared Error" and "Coefficient of determination" will test the performance of our Machine Learning model
```

### What's `Ordinary Least Squares regression`?

**ELI5:** A way of using data to find an equation that explains how two things are related.

**Also,** The equation it finds is a `model`.

**MORE...** For example, if we have data about how much it rains and how much coffee people buy in a town, you could use OLS regression to find the model that best fits the data. The model will tell you how much coffee people will buy for each inch of rain that falls.

**lvlUp:** Ordinary Least Squares regression (OLS) is a common technique for estimating coefficients of linear regression equations which describe the relationship between one or more independent quantitative variables and a dependent variable (simple or multiple linear regression).
