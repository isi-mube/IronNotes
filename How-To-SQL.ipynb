### What is SQL?
<a href=https://stackoverflow.com/questions/2938908/what-is-structured-in-structured-query-language class="button" role="button" style="text-align:center;color:blue;">Stackoverflow</a> is your friend.

In short, SQL stood for "Structured English Query Language" (SEQUEL), meant to do queries that everyone could read. Later renamed. <a href="https://web.archive.org/web/20181122134941/http://www.almaden.ibm.com/cs/people/chamberlin/sequel-1974.pdf">SEQUEL</a> 1970 documentation.

**Why** structured? 

Well, you can **only** use structured English:
* You can do > `select` `col1` `from` `table1`
* but **no** > `give` `col1` `out` `of` `table`

### How's the typical structure?

&darr; **SELECT** , followed by the columns you wish to retrieve. `*` for all columns.

&darr; **FROM** , to specify the table from which you want to retrieve the data.

&darr; **WHERE** , to filter based on specific criteria, we can use conditions.

&darr; **GROUP BY** , optional)to group the data by column/s.

&darr; **HAVING** , typically used after GROUP BY, to filter data based on condition.

&darr; **ORDER BY** , how you want to sort your columns; `ASC` or `DESC`

### **How-To Operate...?**
_____
**Comparision** operators:
* **=** - Compares if two values are equal.
* **<>** or **!=** - Compares if two values are not equal.
* **>** or **<** - Compares if one value is greater or less than another value.
* **>=** - Compares if one value is greater than or equal to another value.
* **<=** - Compares if one value is less than or equal to another value.
_____
About **Logical** operators...
* **AND** - Combines two or more conditions and returns true if all conditions are true.
* **OR** - Combines two or more conditions and returns true if at least one condition is true.
* **NOT** - Reverses the truth value of a condition.
_____
About **Other** operators...
* **BETWEEN** - if a value is within a specified range.
* **IN** - if a value is within a list of specified values.
* **LIKE** - if a value matches a specified pattern.
* **IS NULL** - if a value is NULL.
* **IS NOT NULL** - if a value is not NULL.
_____

### **How-To! Calculate...?**

About **Aggregate Functions**...

### **How-To! Join Tables...?**
_____

About **JOIN Types**...
* **INNER JOIN** - Rows where there is a match in both tables.
* **LEFT JOIN** (or **LEFT OUTER JOIN**) - All rows from the left table and the matched rows from the right table.
* **RIGHT JOIN** (or **RIGHT OUTER JOIN**) - All rows from the right table and the matched rows from the left table. 
* **FULL JOIN** (or **FULL OUTER JOIN**) - All rows when there is a match in either the left or right table. 

**IF NO** match, NULL values are returned for the table with no matching row.

<div class="alert alert-block alert-info">About 
<b>Joining Tables...</b>
When joining tables, use the <b>ON</b> keyword followed by the matching condition, typically using the primary_key and foreign_key relationship between the tables.</div>

About **Keys in Database**...
* **Primary Key** - Unique identifier for each row in a table.
* **Foreign Key** - Column(s) in one table referring to the primary key in another table.
